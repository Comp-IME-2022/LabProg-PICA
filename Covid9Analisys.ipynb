```python
import pandas as pd
import numpy as np
from datetime import datetime
import matplotlib.pyplot as plt
from datetime import datetime
%matplotlib inline
```

```python
#Carregando o dataset
df = pd.read_csv("caso_full.csv")
#Pré-processamento
df['date'] = list(map(lambda x: datetime.strptime(x, '%Y-%m-%d'), df['date']))
df.set_index('date', inplace=True)
```

```python
def gerarGraficoPorEstadoCasos(df):
    estados = [
        'AC','AL','AP','AM','BA','CE','DF','ES','GO','MA','MT','MS','MG','PA','PB','PR','PE','PI','RJ','RN','RS','RO','RR','SC','SP','SE','TO'
    ]
    for estado in estados:
        #Filtragem das informações
        newDf = df[df['state'] == estado]
        newDf = newDf[newDf['place_type'] == 'state']
        newDf['media_movel_casos'] = newDf.iloc[:,15].rolling(window = 7).mean()
        
        #Plotar o gráfico 
        newDf['media_movel_casos'].plot(title= "Média móvel de casos " + estado)
        plt.show()
        
def gerarGraficoPorEstadoObitos(df):
    estados = [
        'AC','AL','AP','AM','BA','CE','DF','ES','GO','MA','MT','MS','MG','PA','PB','PR','PE','PI','RJ','RN','RS','RO','RR','SC','SP','SE','TO'
    ]
    for estado in estados:
        #Filtragem das informações
        newDf = df[df['state'] == estado]
        newDf = newDf[newDf['place_type'] == 'state']
        newDf['media_movel_obitos'] = newDf.iloc[:,16].rolling(window = 7).mean()
        
        #Plotar o gráfico 
        newDf['media_movel_obitos'].plot(title= "Média móvel de óbitos " + estado)
        plt.show()
```

```python
gerarGraficoPorEstadoObitos(df)
```
